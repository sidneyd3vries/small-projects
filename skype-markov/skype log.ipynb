```python
import re
```

```python
with open('gdd + names.txt', 'r') as f:
    raw = [line for line in f]
f.close()
raw[:10]
```

```python
# Remove image/file tags
data = [string for string in raw if '<URIObject' not in string]
data = [string for string in data if "<file" not in string]
```

```python
# Remove hyperlinks
data = [string for string in data if 'http' not in string]
```

```python
# Remove quotation marks
quotation = re.compile(r'\"')
data = [re.sub(quotation, '', string) for string in data]
```

```python
# Replace ascii codes
apos = re.compile(r'&apos;')
data = [re.sub(apos, "'", string) for string in data]
quot = re.compile(r'&quot;')
data = [re.sub(quot, "\"", string) for string in data]
gt = re.compile(r'&gt;')
data = [re.sub(gt, '>', string) for string in data]
amp = re.compile(r'&amp;')
data = [re.sub(amp,'&', string) for string in data]
```

```python
# Remove encoding errors
data = [string.decode('latin-1') for string in data]
```

```python
# Remove newlines and tabs
tab = re.compile(r'\t')

data = [re.sub(tab, ']: ', string) for string in data]

# Add name brackets
data = ['[' + string for string in data]
```

```python
# Remove emoji tags
tag = re.compile(r'<.*?>')
data = [re.sub(tag, '', i) for i in data]
```

```python
sentences = [unicode(string).encode('utf-8') for string in data if len(string.split()) > 5]
len(sentences)
```

```python
with open('out.txt', 'w') as f:
    [f.write("%s\n" % i) for i in sentences[len(sentences)/2:]]
f.close()
```

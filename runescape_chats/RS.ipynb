```python
df1text = df1['msg.value'].tolist()
df2text = df2['msg.value'].tolist()
```

```python
def make_sen(l):
    stop = ['?', '.', '!']
    out = []
    for word in l:
        if isinstance(word, basestring) and word[-1] not in stop:
            out.append(word.strip() + '.')
    return out            
```

```python
clean1 = ' '.join(make_sen(df1text)).decode('utf-8')
```

```python
clean2 = ' '.join(make_sen(df2text)).decode('utf-8')
```
